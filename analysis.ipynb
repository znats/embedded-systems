```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

aapl_candlesticks = pd.read_csv(f'{base_dir}AAPL_candlesticks.txt', header=None, names=['minute', 'open', 'high', 'low', 'close', 'volume'])

aapl_candlesticks['minute'] = pd.to_datetime(aapl_candlesticks['minute'])

fig, ax = plt.subplots()

ax.plot(aapl_candlesticks['minute'], aapl_candlesticks['open'], label='Open Price')
ax.plot(aapl_candlesticks['minute'], aapl_candlesticks['high'], label='High Price')
ax.plot(aapl_candlesticks['minute'], aapl_candlesticks['low'], label='Low Price')
ax.plot(aapl_candlesticks['minute'], aapl_candlesticks['close'], label='Close Price')

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
plt.gcf().autofmt_xdate()

plt.xlabel('Time')
plt.ylabel('Price')
plt.title('AAPL Candlestick Data')
plt.legend()
plt.show()
```

```python
aapl_moving_avg = pd.read_csv(f'{base_dir}AAPL_moving_avg.txt', header=None, names=['minute', 'price_moving_avg', 'volume_moving_avg'])

aapl_moving_avg['minute'] = pd.to_datetime(aapl_moving_avg['minute'])

plt.figure(figsize=(10, 6))
plt.plot(aapl_moving_avg['minute'], aapl_moving_avg['price_moving_avg'], label='15-Minute Price Moving Average')

plt.xlabel('Time')
plt.ylabel('Price Moving Average')
plt.title('AAPL 15-Minute Moving Average')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(aapl_candlesticks['minute'], aapl_candlesticks['close'], label='Close Price', alpha=0.7)

ax.plot(aapl_moving_avg['minute'], aapl_moving_avg['price_moving_avg'], label='15-Minute Price Moving Average', color='orange')

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
plt.gcf().autofmt_xdate()

plt.xlabel('Time')
plt.ylabel('Price')
plt.title('AAPL Candlestick and Moving Average')
plt.legend()
plt.show()
```

```python
import pandas as pd
from datetime import datetime

aapl_trades = pd.read_csv('AAPL_trades.txt', header=None, names=['timestamp', 'price', 'volume'])

aapl_trades['timestamp'] = pd.to_datetime(aapl_trades['timestamp'], unit='ms')

aapl_trades['processing_time'] = aapl_trades['timestamp'].dt.floor('min') + pd.Timedelta('59 seconds')

aapl_trades['time_diff'] = (aapl_trades['processing_time'] - aapl_trades['timestamp']).dt.total_seconds() * 1000

aapl_trades[['timestamp', 'time_diff']].to_csv('AAPL_time_diffs.txt', index=False)
```
